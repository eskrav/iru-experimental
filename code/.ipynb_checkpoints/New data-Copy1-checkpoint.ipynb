```python
import pandas as pn
import regex as re
import os
import csv
import glob
import statistics as s
import numpy as np
from fuzzywuzzy import fuzz
import hashlib
```

```python
filenames = glob.glob("/Users/ekravtch/Shared/informationally-redundant-utterances/data/MTurk/raw/*",recursive=True)
```

```python
files = []

for x in filenames:
    files.append(pn.read_csv(x))

data = pn.concat(files)
data = data.reset_index(drop=True)
data
```

```python
data = data.rename(columns={"WorkerId":"id","SecondsOnHIT":"time"}) 
cols = ["id","time","trials","demographics"]
data = data[cols]
data = data.replace({'\[{':'','{':'','}\]':''}, regex=True)
data = data[~((data["id"] == "AOFME7MQ29C1Q") & (data["time"] == 288))]
data
```

```python
print(f'Average: {round(s.median(data["time"])/60,2)} minutes')
```

```python
s = data["trials"].str.split('},').apply(pn.Series, 1).stack()
s.index = s.index.droplevel(-1)
s.name = "trials"
s
```

```python
ratings = data[["id","time"]]

ratings = ratings.join(s.apply(lambda x: pn.Series(x.split('","'))))

ratings
```

```python
ratings = ratings[["id",0,1,2,3,4,5,6,7]]
ratings = ratings.rename(columns={0:"story",1:"condition",2:"experiment",3:"predictable",4:"unpredictable",5:"script",6:"interaction",7:"rt"})
ratings = ratings.replace({'"story":"':'','prag_condition":"':'','experiment":"':'','redundantQ":"':'','informativeQ":"':'','distractorQ1":"':'','distractorQ2":"':'','rt":':'','^a_':'','^b_':'','^c_':'','^d_':'','^e_':'','^f_':'','^informative':'wonky','biased':'wonky','informative$':'unpredictable','redundant$':'predictable','1_filler':'filler1','2_filler':'filler2','3_filler':'filler3','4_filler':'filler4'}, regex=True)
ratings["trial"] = list(range(1,11))*int(len(ratings.index)/10)
ratings['world'], ratings['activity'] = ratings['condition'].str.split('_', 1).str


ratings = pn.melt(ratings, id_vars=['id', 'trial','story','world','activity','experiment','rt'], value_vars=['predictable', 'unpredictable','script','interaction'])
ratings
```

```python
ratings['condition'] = np.where(ratings['activity'].isin(['predictable', 'unpredictable']), 'after', 'before')
ratings = ratings.rename(columns={"variable":"Qtype","value":"rating"})

ratings
```

```python
z = data["demographics"].str.split('},').apply(pn.Series, 1).stack()
z.index = z.index.droplevel(-1)
z.name = "demographics"
z
```

```python
demographics = data[["id","time"]]

demographics = demographics.join(z.apply(lambda x: pn.Series(x.split('","'))))

demographics
```

```python
demographics = demographics[["id",0,1,2,3,4]]
demographics = demographics.rename(columns={0:"gender",1:"age",2:"language",3:"comments",4:"thoughts"})
demographics = demographics.replace({'"gender":"':'','^age":"':'','language":"':'','comments":"':'','about":"':'','"}':''}, regex=True)

demographics
```

```python
# demographics["language"].str.contains("English")
demographics["language"] = demographics["language"].replace([' $','\n','\t'],'',regex=True).str.lower()
demographics["eligible"] = (demographics.apply(lambda s: fuzz.ratio(s['language'], "english"), axis=1)>80) & (demographics["age"]!="17") & (demographics["age"]!="66")
demographics
```

```python
finalRatings = pn.merge(ratings, demographics[["id","eligible"]], how='left', on="id")
finalRatings
```

```python
demographics = pn.merge(demographics, finalRatings[["id","experiment"]], how='left', on="id").drop_duplicates()
```
