```python
import pandas as pn
import regex as re
import csv
import os
import statistics as s
import numpy as np
```

```python
i = 1
files = []

while os.path.exists(f"/Users/ekravtch/Shared/informationally-redundant-utterances/data/MTurk/plain_{i}.csv"):
    files.append(pn.read_csv(f"/Users/ekravtch/Shared/informationally-redundant-utterances/data/MTurk/plain_{i}.csv"))
    i += 1
data = pn.concat(files)
```

```python
data = data.rename(columns={"WorkerId":"id","SecondsOnHIT":"time"}) 
cols = ["id","time","trials","demographics"]
data = data[cols]
data = data.replace({'\[{':'','{':'','}\]':''}, regex=True)
data
```

```python
print(f'Average: {round(s.mean(data["time"])/60,2)} minutes')
```

```python
data["trials"][0]
```

```python
# split into trials
# remove extraneous marks (brackets, braces, quotes)
# get rid of irrelevant fields (["marker1","marker2"])
# transpose those things into columns by participant, with columns:
# experiment, trial, story, condition (before/after), rt, context, utterance, Qtype, rating

# save into data data frame

# data["trials"][0].split('},{')[0].split(',')

# data["trials"] = data["trials"].replace({'[{':''})
s = data["trials"].str.split('},').apply(pn.Series, 1).stack()
s.index = s.index.droplevel(-1)
s.name = "trials"
s
```

```python
ratings = data[["id","time"]]

ratings = ratings.join(s.apply(lambda x: pn.Series(x.split('","'))))

ratings
```

```python
ratings = ratings[["id",0,1,2,5,6,7,8,9]]
ratings = ratings.rename(columns={0:"story",1:"condition",2:"experiment",5:"predictable",6:"unpredictable",7:"script",8:"interaction",9:"rt"})
ratings = ratings.replace({'"story":"':'','prag_condition":"':'','experiment":"':'','redundantQ":"':'','informativeQ":"':'','distractorQ1":"':'','distractorQ2":"':'','rt":':'','^a_':'','^b_':'','^c_':'','^d_':'','^e_':'','^f_':'','^informative':'wonky','biased':'wonky','informative$':'unpredictable','redundant$':'predictable'}, regex=True)
ratings["trial"] = list(range(1,7))*int(len(ratings.index)/6)
ratings['world'], ratings['activity'] = ratings['condition'].str.split('_', 1).str


ratings = pn.melt(ratings, id_vars=['id', 'trial','story','world','activity','experiment','rt'], value_vars=['predictable', 'unpredictable','script','interaction'])
ratings
```

```python
ratings['condition'] = np.where(ratings['activity'].isin(['predictable', 'unpredictable']), 'after', 'before')
ratings
```

```python
z = data["demographics"].str.split('},').apply(pn.Series, 1).stack()
z.index = z.index.droplevel(-1)
z.name = "demographics"
z
```

```python
demographics = data[["id","time"]]

demographics = demographics.join(z.apply(lambda x: pn.Series(x.split('","'))))

demographics
```

```python
demographics = demographics[["id",0,1,2,3,4]]
demographics = demographics.rename(columns={0:"gender",1:"age",2:"language",3:"comments",4:"thoughts"})
demographics = demographics.replace({'"gender":"':'','^age":"':'','language":"':'','comments":"':'','about":"':'','"}':''}, regex=True)

demographics
```

```python
ratings.to_csv("/Users/ekravtch/Shared/informationally-redundant-utterances/data/MTurk/plain.csv", index=False)
```
